```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

Grafica super sencilla

```python
x=[2,4,6]
y=[2,6,9] 
```

```python
plt.plot(x,y) #grafica sencilla, linea
plt.show()
```

```python
plt.bar(x,y) #grafica de barras
plt.show()
```

```python
plt.scatter(x,y) #Grafico de coordenadas
plt.show()
```

```python
estados = pd.read_excel('PobEstados.xlsx')
```

```python
estados
```

```python
plt.barh(estados['Estado'],estados['Mujeres']) #Grafica de barras de la poblacion de mujeres por estado
plt.title('Mujeres por estado') #Poner titulo a la grafica
plt.ylabel('Estados')
plt.xlabel('Numero de mujeres')
```
