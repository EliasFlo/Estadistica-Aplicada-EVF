a) calcula la mediana de la cantidad que pidieron prestado (amount of money borrowed), asigna un 1 a quienes se encuentran igual o arriba de la mediana (deben mucho) y 0 a los que se encuentran debajo de la mediana (deben poco)

b) Elabora y grafica juntas las funciones de supervivencia por el método de Kaplan-Meier para los que deben mucho y para los que deben poco.

c)Interpreta las graficas de supervivencia, ¿qué significan? ¿quién paga más rápido? ¿quién tarda en pagar más?

```python
#a)
mediana = datos['amount'].median()
```

```python
datos['deben_mucho'] = datos['amount'] > mediana 
datos['deben_poco'] = datos['amount'] < mediana 
datos
```

```python
#b)
debenmucho = datos[datos['deben_mucho'] == True]
debenpoco = datos[datos['deben_poco'] == True]
kmf1 = KaplanMeierFitter(label="Deben mucho")
kmf1.fit(debenmucho['duration'], debenmucho['full_repaid'])

kmf2 = KaplanMeierFitter(label="Deben poco")
kmf2.fit(debenpoco['duration'],debenpoco['full_repaid']) 

kmf1.plot();
kmf2.plot();
```

c)
Interpretación: Lo que la grafica me dice es que los que deben poco, tardan menos en pagar y los que deben mucho tardan mas en pagar

```python
#d)
from lifelines.statistics import logrank_test
```

```python
results = logrank_test(debenmucho['duration'], debenpoco['duration'], debenmucho['full_repaid'], debenpoco['full_repaid'])
results.print_summary()
```
