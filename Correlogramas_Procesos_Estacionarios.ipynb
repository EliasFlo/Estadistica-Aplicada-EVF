## AR(1) vs MA(1) vs ARMA(1,1)

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

#librerias de series de tiempo
from statsmodels.tsa.stattools import adfuller #dickey-fuller
from statsmodels.graphics.tsaplots import plot_acf #grafica de autocorrelacion
from statsmodels.graphics.tsaplots import plot_pacf #grafica de autocorrelacion parcial
from statsmodels.tsa.arima_process import ArmaProcess #medias moviles

%matplotlib inline 

#no salgan advertencias
import warnings
warnings.filterwarnings('ignore')
```

```python
## AR(1) 0.85
np.random.seed(228)

ma0 = np.array([1])
ar1 = np.array([1, -0.85])
proceso_ar1 = ArmaProcess(ar1,ma0).generate_sample(950)
```

```python
fig, ax= plt.subplots(1,2, figsize=(12,4))
#Primer subgrafica
plot_acf(proceso_ar1, zero=False, lags=20, ax=ax[0],title='Autocorrelación ACF')
ax[0].set_ylim(-1,1)
#Segunda subgrafica
plot_pacf(proceso_ar1, zero=False, lags=20, ax=ax[1],title='Autocorrelación PACF')
ax[1].set_ylim(-1,1)

plt.show
plt.savefig('CorrAr1.png')
```

```python
## MA(1) 0.85
np.random.seed(228)

ma1 = np.array([1, 0.85])
ar0 = np.array([1])
proceso_ma1_85 = ArmaProcess(ar0,ma1).generate_sample(950)
```

```python
fig, ax= plt.subplots(1,2, figsize=(12,4))
#Primer subgrafica
plot_acf(proceso_ma1_85, zero=False, lags=20, ax=ax[0],title='Autocorrelación ACF')
ax[0].set_ylim(-1,1)
#Segunda subgrafica
plot_pacf(proceso_ma1_85, zero=False, lags=20, ax=ax[1],title='Autocorrelación PACF')
ax[1].set_ylim(-1,1)

plt.show
plt.savefig('CorrMa1_85.png')
```

```python
#ARMA(1) 0.85, 0.85
np.random.seed(228)

ma1 = np.array([1, 0.85])
ar1 = np.array([1, -0.85])
proceso_arma_85 = ArmaProcess(ar1,ma1).generate_sample(950)
```

```python
fig, ax= plt.subplots(1,2, figsize=(12,4))
#Primer subgrafica
plot_acf(proceso_arma_85, zero=False, lags=20, ax=ax[0],title='Autocorrelación ACF')
ax[0].set_ylim(-1,1)
#Segunda subgrafica
plot_pacf(proceso_arma_85, zero=False, lags=20, ax=ax[1],title='Autocorrelación PACF')
ax[1].set_ylim(-1,1)

plt.show
plt.savefig('CorrArma_85.png')
```
