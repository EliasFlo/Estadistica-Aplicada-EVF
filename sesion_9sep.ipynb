```python
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import expon
import matplotlib.pyplot as plt
```

### Hicimos una variable que representa el tiempo $t$ en el intervalo 0,10 a 100partes

```python
t = np.linspace(0,10,100)
```

```python
t
```

## Exponencial

### Funcion de supervivencia de la exponencial

```python
plt.plot(t,expon.sf(t))
plt.title('Funcion de Supervivencia Exponcial \n $\lambda=1$')
plt.ylabel('S(t)')
plt.xlabel('T')
plt.xlim(0,10)
plt.show()
```

### Funcion de riesgo de la exponencial

```python
plt.plot(t,expon.pdf(t)/expon.sf(t))
plt.title('Funcion de Riesgo Exponcial \n $\lambda=1$')
plt.ylabel('h(t)')
plt.xlabel('T')
plt.xlim(0,10)
plt.show()
```

## Weibull

```python
from scipy.stats import exponweib
```

```python
gamma = 0.5
```

### Funcion de supervivencia de la Weibull

```python
plt.plot(t,exponweib.sf(t,gamma,1))
plt.title(f'Funcion de Supervivencia Weibul \n $\lambda=1$ , $\gamma={gamma}$')
plt.ylabel('S(t)')
plt.xlabel('T')
plt.xlim(0,10)
plt.show()
```
